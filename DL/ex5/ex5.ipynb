# Vector Operations 

ex1(1) - (3)

```python
import torch
import numpy as np

# Exercise 1
# 1.1 Linear function
alpha = 1 / 6
x = torch.tensor([
    [-2],
    [9],
    [3],
    [-4]
])
b = torch.tensor([
    [1],
    [1/2],
    [-1/2],
    [-2]
])

y = alpha * x + b
print(f"y Linear Function: {y}")

# 1.2 Dot product
x = torch.tensor(
    [1, -2, -0.5, 3],
    dtype=torch.float32
)
y = torch.tensor(
    [-3, 7, 2, 2],
    dtype=torch.float32
)
z = torch.dot(x, y)
print(f"z Dot Product: {z}")

# 1.3 Hadamard product
x = torch.tensor(
    [1, 0, 0.5],
    dtype=torch.float32
)
y = torch.tensor(
    [1, 2, 8],
    dtype=torch.float32
)
z = x * y
print(f"z Hadamard Product: {z}")


```

# Vector Similarity

ex2 (1) (2)

```python

# Exercise 2 Comput the pair-wise cosine similarity between
# (1) dog and puppy
# (2) dog and car

dog = torch.tensor(
    [2, 1.5, 3],
    dtype=torch.float32
)
puppy = torch.tensor(
    [3.1, 1.25, 4],
    dtype=torch.float32
)
car = torch.tensor(
    [-4.7, 2.1, -5.5],
    dtype=torch.float32
)

similarity_dog_puppy = torch.dot(dog, puppy) / (torch.norm(dog) * torch.norm(puppy))
similarity_dog_car = torch.dot(dog, car) / (torch.norm(dog) * torch.norm(car))
print(f"similarity_dog_puppy: {similarity_dog_puppy}")
print(f"similarity_dog_car: {similarity_dog_car}")

# Alternatively
similarity_dog_puppy = torch.cosine_similarity(dog, puppy, dim=0)
similarity_dog_car = torch.cosine_similarity(dog, car, dim=0)
print(f"similarity_dog_puppy (Torch): {similarity_dog_puppy}")
print(f"similarity_dog_car (Torch): {similarity_dog_car}")
    
```

$ (3) \; C = AB \text{ and } D = BA \text{ for } 
A = \begin{bmatrix} 1 & 2 & -1 \\ 2 & -2 & \tfrac{1}{2} \end{bmatrix},
\;
B = \begin{bmatrix} 2 & 3 \\ 0 & 1 \\ -2 & -1 \end{bmatrix} $

```python
# Exercise 3 - Compute Matrix Operations
# 3.1
A = torch.tensor(
    [[3, 1, 0],
    [-1, -2, 5]],
    dtype=torch.float32
)
print(A.T)

# 3.2 
# Wx + b
W = torch.tensor(
    [[1, 2, -1],
    [2, -2, 1/2]],
    dtype=torch.float32
)
x = torch.tensor(
    [2, 0, -1],
    dtype=torch.float32
)
b = torch.tensor(
    [0, 1],
    dtype=torch.float32
)
print(f"W @ x + b: {W @ x + b}")

# 3.3
A = torch.tensor(
    [[1, 2, -1],
    [2, -2, 1/2]],
    dtype=torch.float32
)
B = torch.tensor(
    [[2, 3],
    [0, 1],
    [-2, -1]],
    dtype=torch.float32
)

C = A @ B
D = B @ A

print(f"C: {C}")
print(f"D: {D}")
```
