### Vector operations in PyTorch

```python
# x + scalar, - x /
#[1, 2] + 2 = [3, 4]
row = torch.tensor([1, 2])
out = row + 2
print(out)
```

```python
# x + y, - x /
# [1, 2] + [3, 4] = [4, 6]
x = torch.tensor([[1, 2]]) # 1, n
y = torch.tensor([
    [3], 
    [4]]) # 1, n
out = x + y
print('x', x, x.shape)
print('y', y, y.shape)
print('out', out, out.shape)
# broadcasting
# [1, 2]
# [1, 2]

# 3  3
# 4  4

# 4, 5
# 5, 6
```

### Operations: Addition and Subtraction

```python
# matrix + scalar
mat = torch.tensor([[1, 2], 
       [2, 3]])
print('mat', mat)
out = mat + 1
print('out', out)

```

```python
# matrix + vector? 
# broadcast (opt)
mat = torch.tensor([
    [2, 2],
    [3, 3]
])
vec = torch.tensor([[2, 2]])
print('mat', mat, mat.shape)
print('vec', vec, vec.shape)
# [2, 2]
# [2, 2]
print(mat - vec)
```

```python
# matrix + matrix
mat1 = torch.tensor(
    [
        [1, 2],
        [4, 10]
    ]
)
mat2 = torch.tensor(
    [
        [1, 2],
        [2, 5]
    ]
)
print(mat1.shape)
print(mat2.shape)
print(mat1 - mat2)
```

### Operations: multiplication

```python
# x*y [1, 2] * [2, 3]
# broadcast (opt)
# non-singleton dim (opt)

import torch
matx = torch.tensor(
    [
        [15, 25] # row vector or tensor
    ]
)
print('matx', matx, matx.shape)

maty = torch.tensor(
    [
        [1, 1, 1],  # row 1
        [2, 2, 2]   # row 2
    ]
)
print('maty', maty, maty.shape)

# torch.matmul()
z = torch.matmul(matx, maty)
# more complete broadcasting than @
print('z with matmul', z, z.shape)
out = matx @ maty
print('out', out, out.shape)
# What's the difference between * and @? Element-wise versus dot product multiplication
# / is element-wise division
```

```python
# x@y
# [m, n] [n, k]
# [3, 2] [2, 1] => [3, 1]
mat = torch.tensor([
    [1, 2],
    [4, 4],
    [5, 5]
])
# [1, 2] * [1, 3]
# 1*1 + 2*3 = 7
vec = torch.tensor(
    [
        [1],
        [3]
    ]
)

out = mat @ vec
print(out.shape)
print(out)
```

```python
# x / y?
# non-singleton dim (opt)
```

```python
# inverse of a matrix
# shape [3, 4]
# print shape
# A * A-1 = I
# yX-1 = WXX-1 --> Identify matrix because its the inverse so it looks like this yX-1 = WX
# W = y / x

x = torch.tensor(
    [
        [1, 2, 3, 4],
        [0, 1, 4, 5],
        [5, 6, 0, 7]
    ],
    dtype=torch.float32
)
print(x)
# y = WX
xTx = x.T@x
# X: 3x4
# XT: 4x3
# XTX: 3x3
print(xTx.inverse())
# Only square matrices have inverses
# yXT = WXXT --> yXT = W * (XXT)
# x + y = 3 and 2x - y = 1
# The solution is the inverse of the formula
idm = xTx @ xTx.inverse()
print(idm)
```
